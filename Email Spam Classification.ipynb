```python
# importing the dependencies for spam email classification
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
```

Data Collection and Preprocessing

```python
# loading the csv file's data to pandas dataframe
email_data = pd.read_csv('/content/email.csv')
```

```python

# printing the first five rows of dataset
email_data.head()
```

```python
# checking the rows and columns of dataset
email_data.shape
```

```python
# checking the missing values in datasets
email_data.isnull().sum()
```

```python
#getting the some information of dataset
email_data.info()
```

```python
# getting the statistics measures of dataset
email_data.describe()
```

Seperating the feature and target

```python
x = email_data['Message']
y = email_data['Category']
```

Splitting the data into train and test data

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=3)
```

```python
print(x.shape,x_train.shape,x_test.shape)
```

```python
vectorizer = CountVectorizer()
```

```python
x_train_vec = vectorizer.fit_transform(x_train)
x_test_vec = vectorizer.transform(x_test)
```

```python
print(x_train_vec)
```

```python
print(x_test_vec)
```

Model Training

```python

# training the model
model = MultinomialNB()
```

```python
model.fit(x_train_vec, y_train)
```

Model Evaluation

```python
# evaluating the score of train dataset
x_train_prediction = model.predict(x_train_vec)
```

```python
# prining the accuracy score of train data
print('Accuracy Score:',accuracy_score(y_train,x_train_prediction))
```

```python
# printing the confusion matrix of training data
print('Confusion matrix:',confusion_matrix(y_train,x_train_prediction))
```

```python
# printing the classification report of training data
print('Classification Report:',classification_report(y_train,x_train_prediction))
```

```python
# evaluatng the score of testing data
x_test_prediction = model.predict(x_test_vec)
```

```python
# printing the accuracy score of testing data
print('Accuracy Score:',accuracy_score(y_test,x_test_prediction))

# printing the confusion matrix of dataaset
print('Confusion Matrix:',confusion_matrix(y_test,x_test_prediction))

# printing the classification report of testing data
print('Classification Report:',classification_report(y_test,x_test_prediction))
```

Predictive System

```python
# input mail
mail = str(input('Enter the mail'))

# vectorize the mail - wrap it in a list
mail_vec = vectorizer.transform([mail])

# make prediction
prediction = model.predict(mail_vec)

# final result
if prediction[0] == 'spam':
    print('The mail is spam')
else:
    print('The mail is not spam')
```
