Importing the depencies.


```python
import numpy as np
import pandas as pd
import difflib
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

Data Collection and preprocessing

```python
movie_data = pd.read_csv('/content/movies_dataset.csv',engine='python',on_bad_lines='skip')
```

```python
movie_data = movie_data.iloc[:,~movie_data.columns.str.contains('Unnamed:')]
```

```python
# printing the first five rows of dataset
movie_data.head()
```

```python
# chcking the number of rows and columns
movie_data.shape
```

```python
# selecting relevent feature for recommendation
selected_feature = ['genres','keywords','tagline','cast','director']
```

```python
print(selected_feature)
```

```python
# replacing the null value with null string
for feature in selected_feature:
  movie_data[feature]=movie_data[feature].fillna('')
```

```python
# combining all the 5 selected feature
combined_feature = movie_data['genres']+' '+movie_data['keywords']+' '+movie_data['tagline']+' '+movie_data['cast']+' '+movie_data['director']
```

```python
print(combined_feature)
```

```python
# converting the text data into feature vectoe
vectorizer = TfidfVectorizer()
```

```python
feature_vector = vectorizer.fit_transform(combined_feature)
```

```python
print(feature_vector)
```

Cosine Similarity

```python
# getting the similarity values using cosine similarity
similarity = cosine_similarity(feature_vector)
```

```python
print(similarity)
```

```python
print(similarity.shape)
```

```python
# getting the movie name
movie_name = input('enter your favourite movie name: ')
```

```python
# creating a list  with all the movie name in given dataset
list_of_all_titles = movie_data['title'].dropna().astype(str).tolist()
```

```python
print(list_of_all_titles)
```

```python
movie_name = str(movie_name)
```

```python
find_close_match = difflib.get_close_matches(movie_name,list_of_all_titles)
```

```python
print(find_close_match)
```

```python
close_match = find_close_match[0]
```

```python
print(close_match)
```

```python
# finding the index with moviee's title
index_of_movie = movie_data[movie_data.title==close_match]['index'].values[0]
```

```python
print(type(index_of_movie))
```

```python
index_of_movie = int(index_of_movie)
```

```python
print(index_of_movie)
```

```python
# getting a list of similar movies
similarity_score = list(enumerate(similarity[index_of_movie]))
```

```python
print(similarity_score)
```

```python
len(similarity_score)
```

```python
# sorting the movies based on their similarity score
sorted_similar_movies = sorted(similarity_score,key=lambda x:x[1],reverse=True)
```

```python
print(sorted_similar_movies)
```

```python
# print the name of similar movies based on index
print('Movies suggested for you:\n')

i = 1

for movie in sorted_similar_movies:
    index = movie[0]  # movie = (index, score)

    # Get the movie title from the DataFrame
    title_from_index = movie_data[movie_data.index == index]['title'].values[0]
    title_from_index = str(title_from_index)


    if i < 30:
        print(i, '.', title_from_index)
        i += 1
```
